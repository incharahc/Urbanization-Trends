# 4. Dataset Laoding and Cleaning





1. **Loading Datasets**:

* Loads all three datasets using pandas.read_csv.
* Adjusts paths to ensure the correct files are accessed.

2. **Cleaning the Main Dataset**:

* Drops unnecessary columns (e.g., "Unnamed: 66").
* Renames columns for better readability.
* Handles missing values using forward fill and backward fill methods or drops rows with all NaN values (Ye et al. 2023).

3. **Cleaning Metadata**:

* Renames columns in country and indicator metadata.
* Fills missing values with placeholders like "Unknown".
    
4. **Merging Datasets**:

* Merges the main dataset with country metadata to include region and income group information.
    
5. **Exporting Cleaned Data**:

* Exports the cleaned and merged dataset to a CSV file for further analysis.

    

```python




# Import necessary libraries
import pandas as pd
import numpy as np

# File paths 
main_data_path = 'API_SP.URB.TOTL.IN.ZS_DS2_en_csv_v2_972.csv'
country_metadata_path = 'Metadata_Country_API_SP.URB.TOTL.IN.ZS_DS2_en_csv_v2_972.csv'
indicator_metadata_path = 'Metadata_Indicator_API_SP.URB.TOTL.IN.ZS_DS2_en_csv_v2_972.csv'

# Load the main dataset
main_data = pd.read_csv(main_data_path, skiprows=4)

# Load the country metadata
country_metadata = pd.read_csv(country_metadata_path)

# Load the indicator metadata
indicator_metadata = pd.read_csv(indicator_metadata_path)

# Display the first few rows of the datasets to understand their structure
print("Main Dataset:")
print(main_data.head())
print("\nCountry Metadata:")
print(country_metadata.head())
print("\nIndicator Metadata:")
print(indicator_metadata.head())

# ----------- DATA CLEANING -----------

# Step 1: Clean the Main Dataset
# Drop unnecessary columns
columns_to_drop = ['Unnamed: 66', 'Unnamed: 68']  # Adjust this if more columns are unnecessary
main_data = main_data.drop(columns=[col for col in columns_to_drop if col in main_data.columns])

# Rename columns for better readability
main_data.rename(columns={"Country Name": "Country", "Country Code": "Country_Code", "Indicator Name": "Indicator", "Indicator Code": "Indicator_Code"}, inplace=True)

# Check for missing values and handle them
print("\nMissing values in the main dataset:")
print(main_data.isnull().sum())

# Drop rows where all years have NaN values
main_data.dropna(how='all', subset=main_data.columns[4:], inplace=True)

# Forward fill and backward fill for missing values in the main dataset
main_data.ffill(inplace=True)  # Forward fill
main_data.bfill(inplace=True)  # Backward fill

# Reset index after cleaning
main_data.reset_index(drop=True, inplace=True)

# Step 2: Clean the Country Metadata
# Rename columns for better readability
country_metadata.rename(columns={"Country Code": "Country_Code", "Region": "Region", "IncomeGroup": "Income_Group"}, inplace=True)

# Handle missing values in the metadata
print("\nMissing values in the country metadata:")
print(country_metadata.isnull().sum())

# Fill missing region or income group values explicitly
country_metadata['Region'] = country_metadata['Region'].fillna("Unknown")
country_metadata['Income_Group'] = country_metadata['Income_Group'].fillna("Unknown")

# Step 3: Clean the Indicator Metadata
# Rename columns for better readability
indicator_metadata.rename(columns={"INDICATOR_CODE": "Indicator_Code", "INDICATOR_NAME": "Indicator"}, inplace=True)

# Handle missing values in the indicator metadata
print("\nMissing values in the indicator metadata:")
print(indicator_metadata.isnull().sum())

# Fill missing values explicitly for the indicator metadata
indicator_metadata['Indicator'] = indicator_metadata['Indicator'].fillna("Unknown")

# ----------- MERGING DATASETS -----------

# Merge main data with country metadata to include region and income group
merged_data = pd.merge(main_data, country_metadata[['Country_Code', 'Region', 'Income_Group']], on='Country_Code', how='left')

# Display the merged dataset
print("\nMerged Dataset:")
print(merged_data.head())

# ----------- EXPORT CLEANED DATASET -----------

# Save the cleaned and merged dataset to a CSV file for further use
merged_data.to_csv('cleaned_urbanization_data.csv', index=False)

print("\nCleaned and merged dataset has been saved as 'cleaned_urbanization_data.csv'.")



```

# 5. Exploratory Data Analysis (EDA):





1. **Basic Statistical Summaries**:

* Descriptive statistics for urbanization percentages.
* Missing value check.

2. **Global Urbanization Trends**:

* Line plot showing the global trend in urbanization over time (1960–2023).

3. **Regional Urbanization Trends**:

* Line plots for urbanization trends by region, allowing comparison across different areas.
    
4. **Income Group Urbanization Trends**:

* Line plots for urbanization trends by income group, highlighting differences between high-income, low-income, and middle-income countries.

5. **Top and Bottom 10 Countries**:

* Based on average urbanization, bar plots for the top 10 and bottom 10 countries.

6. **Correlation Analysis**:

* Heatmap showing correlations between yearly urbanization percentages to identify potential patterns or clusters.


    

```python


# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
file_path = 'cleaned_urbanization_data.csv' 
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("Cleaned Dataset:")
print(data.head())

# ----------- BASIC STATISTICAL SUMMARIES -----------

# Summary statistics for urban population percentages
print("\nSummary Statistics for Urban Population (%):")
year_columns = [col for col in data.columns if col.isdigit()]  # Extract year columns (1960-2023)
print(data[year_columns].describe())

# Check for missing values
print("\nMissing Values in the Dataset:")
print(data.isnull().sum())

# ----------- GLOBAL URBANIZATION TRENDS -----------

# Calculate global average urbanization by year
global_trend = data[year_columns].mean().reset_index()
global_trend.columns = ['Year', 'Urbanization (%)']
global_trend['Year'] = global_trend['Year'].astype(int)

# Plot global urbanization trend over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=global_trend, x='Year', y='Urbanization (%)', marker='o')
plt.title('Global Urbanization Trend (1960-2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Urban Population (% of Total)', fontsize=14)
plt.grid()
plt.show()

# ----------- REGIONAL URBANIZATION TRENDS -----------

# Calculate average urbanization by region and year
regional_trends = data.groupby('Region')[year_columns].mean().T
regional_trends.index = regional_trends.index.astype(int)  # Convert index to integer

# Plot regional urbanization trends over time
plt.figure(figsize=(14, 8))
for region in regional_trends.columns:
    plt.plot(regional_trends.index, regional_trends[region], label=region)

plt.title('Regional Urbanization Trends (1960-2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Urban Population (% of Total)', fontsize=14)
plt.legend(loc='upper left', fontsize=10, title='Region')
plt.grid()
plt.show()

# ----------- INCOME GROUP URBANIZATION TRENDS -----------

# Calculate average urbanization by income group and year
income_trends = data.groupby('Income_Group')[year_columns].mean().T
income_trends.index = income_trends.index.astype(int)  # Convert index to integer

# Plot income group urbanization trends over time
plt.figure(figsize=(14, 8))
for group in income_trends.columns:
    plt.plot(income_trends.index, income_trends[group], label=group)

plt.title('Urbanization Trends by Income Group (1960-2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Urban Population (% of Total)', fontsize=14)
plt.legend(loc='upper left', fontsize=10, title='Income Group')
plt.grid()
plt.show()

# ----------- TOP 10 AND BOTTOM 10 COUNTRIES -----------

# Calculate average urbanization for each country over all years
data['Average_Urbanization'] = data[year_columns].mean(axis=1)

# Top 10 countries with highest average urbanization
top_10_countries = data.nlargest(10, 'Average_Urbanization')[['Country', 'Average_Urbanization']]
print("\nTop 10 Countries with Highest Average Urbanization:")
print(top_10_countries)

# Bottom 10 countries with lowest average urbanization
bottom_10_countries = data.nsmallest(10, 'Average_Urbanization')[['Country', 'Average_Urbanization']]
print("\nBottom 10 Countries with Lowest Average Urbanization:")
print(bottom_10_countries)

# Plot top 10 countries with highest average urbanization
plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_10_countries,
    x='Average_Urbanization',
    y='Country',
    color='blue'  # Use a single color for all bars
)
plt.title('Top 10 Countries with Highest Average Urbanization', fontsize=16)
plt.xlabel('Average Urban Population (% of Total)', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Optional: Add gridlines
plt.show()

# Plot bottom 10 countries with lowest average urbanization
plt.figure(figsize=(12, 6))
sns.barplot(
    data=bottom_10_countries,
    x='Average_Urbanization',
    y='Country',
    color='red'  # Use a single color for all bars
)
plt.title('Bottom 10 Countries with Lowest Average Urbanization', fontsize=16)
plt.xlabel('Average Urban Population (% of Total)', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Optional: Add gridlines
plt.show()

# ----------- CORRELATION ANALYSIS -----------

# Calculate correlation matrix for yearly urbanization data
correlation_matrix = data[year_columns].corr()

# Plot heatmap of correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap of Urbanization Data (Yearly)', fontsize=16)
plt.show()



```
